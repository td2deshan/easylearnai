```python
import numpy as np # for array, matrix handing
import matplotlib.pyplot as plt #for data visualization
%matplotlib inline

#change plotting style
#to get all available plotting styles --> plt.style.available
plt.style.use('seaborn')
```

```python
#make dataset
# "reshape(-1, 1)" convert array into a column vector 
temp = np.array([20, 21, 23, 24, 25, 26, 27, 30, 32, 33, 35, 36, 38, 39, 40]).reshape(-1, 1)
sales = np.array([16, 17, 17, 19, 20, 22, 24, 28, 34, 36, 37, 38, 40, 42, 44]).reshape(-1, 1)

#scatter plot dataset
plt.plot(temp, sales, 'o')

plt.title('no of sales vs temp')
plt.xlabel('(avg) temp in celisius')
plt.ylabel('(avg) no of sales')
```

```python
#To make model we need following libraries
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
#Let's make basic linear DL model
#1st -> get object from Sequeantial class
model = Sequential()

#2nd ->  add layers
#in this case only need one simple layer

# Dense -- fully connected layer
# units -- no of outputs
# input_dim -- no of inputs at one time
model.add(Dense(units=1, input_dim=1))

#3rd -> complie created model
# optimizer helps to find optimal values
# lr -- leaning rate
# loss -- calculate error using mean square error(mse)
# metrics -- used to display of progress in each epoch
model.compile(optimizer=Adam(lr=.1), loss='mse', metrics=['accuracy'])
```

```python
#4th -> train our model
#epochs = how many times train model
#verbose = displaying ways(0, 1, 2) at each epoch
model.fit(x=temp, y=sales, epochs=1000, verbose=0)
```

```python
#display what happen to error(loss) at traning model
plt.plot(model.history.history['loss'])

plt.xlabel('no of epoch')
plt.ylabel('error (loss)')
```

```python
#5th -> get prediction using our tranind model
predicted_values = model.predict(x=temp)

#plot predicted line with original points
plt.plot(temp, sales, 'o')
plt.plot(temp, predicted_values)
```

```python
#predict how musch sales happen when temperature is 45 celcius
model.predict([[45]])
```
